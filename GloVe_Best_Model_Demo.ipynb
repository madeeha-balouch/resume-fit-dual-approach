```python
# 1. Get the dataset from Hugging Face
dataset_name = "cnamuangtoun/resume-job-description-fit"
try:
    dataset = load_dataset(dataset_name)
    df = dataset['train'].to_pandas()
    print(f"Dataset '{dataset_name}' loaded successfully. Shape: {df.shape}")
    print("Column names in the DataFrame:")
    print(df.columns)
    # We need to identify the correct column names for text and labels
    TEXT_COLUMN = None  # We will update this based on the output
    LABEL_COLUMN = None # We will update this based on the output
except Exception as e:
    print(f"Error loading dataset '{dataset_name}': {e}")
    exit()
```

```python
# 2. Preprocessing
TEXT_COLUMN_RESUME = 'resume_text'
TEXT_COLUMN_JOB = 'job_description_text'
LABEL_COLUMN = 'label'

stop_words = set(stopwords.words('english'))
custom_stopwords = ['e.g.', 'i.e.', 'etc.', 'the', 'and', 'a', 'an', 'in', 'of', 'to'] # Add more if needed
stop_words.update(custom_stopwords)

def preprocess_text(text):
    if isinstance(text, str):
        # Lowercase
        text = text.lower()
        # Remove email addresses
        text = re.sub(r'\S*@\S*\s?', '', text)
        # Remove non-ASCII characters
        text = re.sub(r'[^\x00-\x7F]+', '', text)
        # Remove numbers (keeping those with % $ or year format)
        text = re.sub(r'(?<!\d)(?<![\$\%])\b\d+\b', '', text)
        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))
        # Remove stopwords
        text_tokens = text.split()
        filtered_tokens = [w for w in text_tokens if not w in stop_words]
        return " ".join(filtered_tokens)
    return ""

print("Applying preprocessing...")
df['processed_resume'] = df[TEXT_COLUMN_RESUME].apply(preprocess_text)
df['processed_job_description'] = df[TEXT_COLUMN_JOB].apply(preprocess_text)
# Combine processed resume and job description for the model
df['combined_text'] = df['processed_resume'] + ' ' + df['processed_job_description']
df = df[df['combined_text'].str.len() > 0] # Remove rows with empty combined text
print("Preprocessing complete.")
```

```python

import nltk
import numpy as np
import gensim.downloader as api
from nltk.tokenize import word_tokenize

# Load GloVe model (pre-trained)
glove_model = api.load("glove-wiki-gigaword-100")  # 100-dimensional GloVe vectors

# Tokenize the texts
df['resume_tokens'] = df['processed_resume'].apply(word_tokenize)
df['job_tokens'] = df['processed_job_description'].apply(word_tokenize)

# Function to get average GloVe embeddings
def average_glove_vectors(tokens, model, vector_size):
    valid_tokens = [token for token in tokens if token in model]
    if not valid_tokens:
        return np.zeros(vector_size)
    return np.mean([model[token] for token in valid_tokens], axis=0)

# Apply to dataframe
df['resume_vec'] = df['resume_tokens'].apply(lambda x: average_glove_vectors(x, glove_model, 100))
df['job_vec'] = df['job_tokens'].apply(lambda x: average_glove_vectors(x, glove_model, 100))

# Prepare training/testing sets
X_resume = np.stack(df['resume_vec'].values)
X_job = np.stack(df['job_vec'].values)
y = df[LABEL_COLUMN]

X_resume_train, X_resume_test, X_job_train, X_job_test, y_train, y_test = train_test_split(
    X_resume, X_job, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Calculate Cosine Similarity
print("Calculating cosine similarity...")

def calculate_glove_similarity(resume_glove, job_glove):
    return cosine_similarity(resume_glove.reshape(1, -1), job_glove.reshape(1, -1))[0][0]

train_similarity = [
    calculate_glove_similarity(X_resume_train[i], X_job_train[i])
    for i in range(X_resume_train.shape[0])
]

test_similarity = [
    calculate_glove_similarity(X_resume_test[i], X_job_test[i])
    for i in range(X_resume_test.shape[0])
]

print("Cosine similarity calculated.")
```

```python
# Prepare data for the model
X_train_glove = pd.DataFrame(X_resume_train, columns=[f'glove_resume_{i}' for i in range(X_resume_train.shape[1])])
X_job_glove = pd.DataFrame(X_job_train, columns=[f'glove_job_{i}' for i in range(X_job_train.shape[1])])
X_train_model = pd.concat([X_train_glove, X_job_glove], axis=1)
X_train_model['similarity'] = train_similarity
X_train_model.columns = X_train_model.columns.astype(str)  # Convert all column names to strings

X_test_glove = pd.DataFrame(X_resume_test, columns=[f'glove_resume_{i}' for i in range(X_resume_test.shape[1])])
X_job_test_glove = pd.DataFrame(X_job_test, columns=[f'glove_job_{i}' for i in range(X_job_test.shape[1])])
X_test_model = pd.concat([X_test_glove, X_job_test_glove], axis=1)
X_test_model['similarity'] = test_similarity
X_test_model.columns = X_test_model.columns.astype(str)    # Convert all column names to strings
```

```python
# Initialize and train the models
gnb = GaussianNB()
gnb.fit(X_train_model, y_train)
gnb_predictions_combined = gnb.predict(X_test_model)


# Evaluate the model with combined features
print("\nGaussian Naive Bayes with GloVe and Cosine Similarity:")
print(classification_report(y_test, gnb_predictions_combined))
print("Accuracy:", accuracy_score(y_test, gnb_predictions_combined))
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_model, y_train)
rf_predictions = rf.predict(X_test_model)

print("\nRandom Forest with GloVe and Cosine Similarity:")
print(classification_report(y_test, rf_predictions))
print("Accuracy:", accuracy_score(y_test, rf_predictions))
```

```python
label_mapping = {'No Fit': 0, 'Potential Fit': 1, 'Good Fit': 2}
y_train_encoded = y_train.map(label_mapping)

y_test_encoded = y_test.map(label_mapping)
```

```python
# Initialize and train the XGBoost model with combined features
xgb_model = xgb.XGBClassifier(random_state=42)
xgb_model.fit(X_train_model, y_train_encoded)  # Use the numerically encoded y_train
xgb_predictions = xgb_model.predict(X_test_model)

# Evaluate the model with combined features
print("\nXGBoost with GloVe Embedding and Cosine Similarity:")
print(classification_report(y_test_encoded, xgb_predictions))
print("Accuracy:", accuracy_score(y_test_encoded, xgb_predictions))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate the confusion matrix
cm = confusion_matrix(y_test_encoded, xgb_predictions)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=xgb_model.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - XGBoost with GloVe Embedding  and Cosine Similarity")
plt.show()
```

```python

def compute_similarity_scores(resume_embed, jd_embed):
    similarities = cosine_similarity(resume_embed.reshape(1, -1), jd_embed.reshape(1, -1))
    return similarities[0, 0]  # each resume vs its own JD



similarity_scores = [
    compute_similarity_scores(X_resume_test[i], X_job_test[i])
    for i in range(X_resume_test.shape[0])
]

mean_similarity = np.mean(similarity_scores)
std_dev_similarity = np.std(similarity_scores)
threshold = mean_similarity + 1 * std_dev_similarity

#xgb_labels = xgb_predictions.map(inverse_label_mapping)
inverse_label_mapping = {v: k for k, v in label_mapping.items()}
xgb_labels = [inverse_label_mapping[pred] for pred in xgb_predictions]


intersection_indices = [
    i for i, (label, score) in enumerate(zip(xgb_labels, similarity_scores))
    if (label == 'Good Fit' or label == 'Potential Fit') and score >= threshold
]

# Prepare color: green for intersection, gray otherwise
colors = ['green' if i in intersection_indices else 'gray' for i in range(len(similarity_scores))]

plt.figure(figsize=(12, 6))
plt.scatter(range(len(similarity_scores)), similarity_scores, c=colors, alpha=0.7)
plt.axhline(y=threshold, color='red', linestyle='--', label=f'Threshold = {threshold}')
plt.xlabel('Resume Index')
plt.ylabel('Semantic Similarity Score')
plt.title('Resume Similarity Scores and Intersection with XGB model Predictions Using GloVe embeddings')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Example inputs
resumes = [
    "I have 5 years of experience with Python and Django backend development.",
    "I am a marketing professional who writes blogs and handles social media.",
    "I studied data science and worked on multiple Python data analysis projects."
]

job_description = "Looking for a backend developer skilled in Python and RESTful API design, preferably with Django experience."

# Optional: use your existing model to get predicted labels
# For this example, let's assume your model already predicted these:
predicted_labels = ["good fit", "no fit", "potential fit"]  # output from xgb_model.predict()

# Step 1: Preprocess + tokenize
resume_tokens = [word_tokenize(preprocess_text(r)) for r in resumes]
job_tokens = word_tokenize(preprocess_text(job_description))

# Step 2: Get GloVe vectors
resume_vecs = [average_glove_vectors(tokens, glove_model, 100) for tokens in resume_tokens]
job_vec = average_glove_vectors(job_tokens, glove_model, 100)

# Step 3: Calculate cosine similarity
cosine_similarities = [
    cosine_similarity(r.reshape(1, -1), job_vec.reshape(1, -1))[0][0]
    for r in resume_vecs
]

# Step 4: Threshold cosine similarity
threshold = 0.5
similarity_pass = [i for i, sim in enumerate(cosine_similarities) if sim >= threshold]

# Step 5: Labels that are good or potential fits
ml_pass = [i for i, label in enumerate(predicted_labels) if label in ("good fit", "potential fit")]

# Step 6: Final intersection
final_indices = set(similarity_pass).intersection(ml_pass)

# Step 7: Output final results
print("Selected resumes based on ML + similarity agreement:")
for i in final_indices:
    print(f"\nResume #{i+1}")
    print("Text:", resumes[i])
    print("Predicted Label:", predicted_labels[i])
    print("Cosine Similarity:", cosine_similarities[i])
```
